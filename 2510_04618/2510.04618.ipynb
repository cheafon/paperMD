# Abstract
ACE提出之前，使用模型权重更新的方式，但是领域见解不好，上下文会稀释，坍塌。而使用ACE无须标记监督，可以实现低开销、高效且自我改进的系统。

# Intro
ACE可以帮助下游任务的提示、回忆过去的事实记忆、减少幻觉，有以下几点重要部分：
1. 给用户和开发者提供可解释性
2. 运行时快速执行新的知识
3. 可以在复杂的模型和模块之间共享
4. 长上下文的技术发展给ACE提供了可能

尽管目前的上下文工程已经有很大进展，但是仍然有以下两个短板：
1. brevity bias：GEPA，广泛适用的prompt优化器，导致领域的启发洞见性不好
2. 上下文collapse坍塌

我们认为上下文不应该是简单的总结，而是一个全面、包容的剧本，可以理解细节、全面、对专业领域有洞察力，和人们不一样，人们只能接受简短的知识，而大模型可以对
长上下文进行关键信息提取和理解。所以上下文不应该是一味的压缩他，把他的领域知识都压缩没了，而是应该让大模型决定在何时决定什么上下文最重要。
所以引出了ACE（Agentic context engineering）。

# ACE



Generator:产生推理轨迹，同时揭露了有效的策略和反复的陷阱。

Reflector：通过这些痕迹提取教训，在多次迭代中进行改进。

Curator：将这些经验教训策展合并成增量的内容。

Delta Context Items：通过非LLM逻辑的确定性的进行上下文的增量更新，避免全量更新的成本和延迟问题，同时确保过去知识保留，以及新知识的稳定追加。

- offline：系统提示词的优化
- 运行时的memory的自适应


# 几种prompt对比
ICL(In-Context Learning):只是简单的喂几个sample，没有反思等过程。

Dymantic Chatsheet:备忘录的全量更新导致高延迟、误差积累从而丢失细节、从而崩溃

CEPA：使用多种策略进行。
